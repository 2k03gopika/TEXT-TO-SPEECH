# Here we are using Python's PYTTSX3 library  Unlike alternative libraries, it works offline, and is compatible with both Python 2 and 3.

# **The pyttsx3 library, which is a Python wrapper for the Microsoft Speech API. First, the engine is initialized and the voice and rate are set. Then, the text to convert is defined and passed to the say() method of the engine object. Finally, the runAndWait() method is called to play the speech.

```python
import pyttsx3
```

# Initialize the text-to-speech engine

```python
def say(sp):
    # Initialize the Text-to-Speech engine
    engine = pyttsx3.init()

    # Set the properties of the voice
    voices = engine.getProperty('voices')
    engine.setProperty('voice', voices[1].id)  # Choose a voice, in this case, the second voice
    engine.setProperty('rate', 150)  # Set the speaking rate, in words per minute

    # Convert text to speech
    engine.say(sp)
    engine.runAndWait()
```

# CODE EXPLANATION:
    The engine object is initialized with the pyttsx3.init() function, which creates a new instance of the Text-to-Speech engine.

The voices variable retrieves the list of available voices that the engine can use. By default, the first voice in the list is selected. However, in this code, the second voice is chosen using voices[1].id. You can change this value to select a different voice.

The rate property sets the speed at which the text is spoken, in words per minute. In this code, the rate is set to 150.

Finally, the engine.runAndWait() function is called to start the speech conversion process and wait until it is finished before the program exits.
